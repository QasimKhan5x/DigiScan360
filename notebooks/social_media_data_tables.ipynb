```python
from pyspark.sql.functions import col
```

```python
user_tweet_df = spark.read.format("delta").load("Tables/user_tweet_data")
user_tweet_df.printSchema()
```

```python
consumer_tweet_df = spark.read.format("delta").load("Tables/consumer_tweet")
consumer_tweet_df.printSchema()
```

```python
facebook_data_df = spark.read.format("delta").load("Tables/facebook_data")
facebook_data_df.printSchema()
```

```python
display(facebook_data_df)
```

```python
# Creating Dim_User table
dim_user = user_tweet_df.select(
    col("id").alias("user_id"),
    col("name"),
    col("username"),
    col("created_at"),
    col("url"),
    col("followers_count"),
    col("record_date")
)
dim_user.show()
dim_user.printSchema()
```

```python
# Creating Dim_Consumer table
dim_consumer = consumer_tweet_df.select(
    col("author_id").alias("author_id"),
    col("screen_name"),
    col("lang"),
    col("brand_name")
)
dim_consumer.show()
dim_consumer.printSchema()
```

```python
# Creating Dim_Facebook table
dim_facebook = facebook_data_df.select(
    col("id").alias("page_id"),
    col("name"),
    col("username"),
    col("contact_address"),
    col("current_location"),
    col("emails")
)
dim_facebook.show()
dim_facebook.printSchema()
```

```python
# Creating Fact_User_Tweet table
fact_user_tweet = user_tweet_df.select(
    col("id").alias("id"),
    col("created_at"),
    col("followers_count"),
    col("record_date"),
    col("friends_count"),
    col("verified")
)
fact_user_tweet.show()
fact_user_tweet.printSchema()
```

```python
# Creating Fact_Consumer_Tweet table
fact_consumer_tweet = consumer_tweet_df.select(
    col("tweet_id").alias("tweet_id"),
    col("author_id"),
    col("in_reply_to_user_id"),
    col("timestamp"),
    col("text"),
    col("impression_count"),
    col("like_count"),
    col("reply_count"),
    col("repost_count"),
    col("quote_count"),
    col("hashtags"),
    col("user_followers_count"),
    col("user_following_count"),
    col("verified")
)
fact_consumer_tweet.show()
fact_consumer_tweet.printSchema()
```

```python
# Creating Fact_Facebook_Interaction table
fact_facebook = facebook_data_df.select(
    col("id").alias("page_id"),
    col("post_id"),
    col("created_time9").alias("created_time"),
    col("fan_count").alias("fans_count"),
    col("followers_count"),
    col("message"),
    col("shares"),
    col("likes").alias("post_like_count"),
    col("like_count").alias("comment_like_count"),
    col("reactions"),
    col("comment_id"),
    col("from").alias("from_user_id"),
    col("comment"),
    col("verified")
)
fact_facebook.show()
fact_facebook.printSchema()
```

```python
display(fact_facebook)
```

```python
# Check if the source DataFrames have true values for the verified column
print("Checking fact_consumer_tweet DataFrame:")
fact_consumer_tweet.filter(fact_consumer_tweet.verified == False).show()

print("Checking fact_user_tweet DataFrame:")
fact_user_tweet.filter(fact_user_tweet.verified == False).show()
```

```python
# Print the schema of the DataFrames to verify the data types
print("fact_consumer_tweet Schema:")
fact_consumer_tweet.printSchema()

print("fact_user_tweet Schema:")
fact_user_tweet.printSchema()
```

```python
delta_table_name = "dim_user"
dim_user.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")

delta_table_name = "dim_consumer"
dim_consumer.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")

delta_table_name = "dim_facebook"
dim_facebook.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")

delta_table_name = "fact_user_tweet"
fact_user_tweet.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")

delta_table_name = "fact_facebook"
fact_facebook.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")

delta_table_name = "fact_consumer_tweet"
fact_consumer_tweet.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable(delta_table_name)
print(f"DataFrame has been written to Delta table: {delta_table_name}")
```

```python
# Inspect data in Delta tables after writing
print("Checking Delta table for fact_consumer_tweet where verified is true:")
spark.sql("SELECT * FROM fact_consumer_tweet WHERE verified = false").show()

print("Checking Delta table for fact_user_tweet where verified is true:")
spark.sql("SELECT * FROM fact_user_tweet WHERE verified = false").show()
```
